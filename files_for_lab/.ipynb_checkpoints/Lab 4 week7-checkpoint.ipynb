```python
import pandas as pd
import numpy as np
import statsmodels.api as sm
```

```python
numerical = pd.read_csv('numerical.csv')
numerical.shape
```

```python
targets = pd.read_csv('target.csv')
targets.shape
```

```python
categorical = pd.read_csv('categorical.csv')
categorical.shape
```

```python
pd.set_option('display.max_columns',None)
numerical.head()
```

```python
pd.set_option('display.max_rows',None)
numerical.dtypes
```

```python
all_data = pd.concat((numerical,categorical),axis=1)
all_data.head()
```

```python
pd.set_option('display.max_rows',None)
categorical.dtypes
```

```python
targets['TARGET_B']
targets['TARGET_B'].value_counts()
#super imbalanced, mostly zero
```

```python
y=targets['TARGET_B']
X=all_data
```

```python
from sklearn.model_selection import train_test_split as tts

X_train, X_test, y_train, y_test=tts(X, y, test_size=.2)
```

```python
train_num = X_train.select_dtypes(include=np.number)
train_num.head()
```

```python
train_cat = X_train.select_dtypes(exclude=np.number)
train_cat.head()
```

```python
test_num = X_test.select_dtypes(include=np.number)
test_cat = X_test.select_dtypes(exclude=np.number)
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler_num = MinMaxScaler()
train_num_scaled = scaler_num.fit_transform(train_num)
train_num_scaled_df = pd.DataFrame(train_num_scaled, columns=train_num.columns, index=train_num.index)
train_num_scaled_df.head()
```

```python
test_num_scaled = scaler_num.transform(test_num)
test_num_scaled_df = pd.DataFrame(test_num_scaled, columns=test_num.columns, index=test_num.index)
test_num_scaled_df.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(sparse_output=False, handle_unknown='ignore')

train_cat_encoded = encoder.fit_transform(train_cat)
test_cat_encoded = encoder.transform(test_cat)

train_cat_encoded_df = pd.DataFrame(train_cat_encoded, columns=encoder.get_feature_names_out(train_cat.columns), index=train_cat.index)
test_cat_encoded_df = pd.DataFrame(test_cat_encoded, columns=encoder.get_feature_names_out(train_cat.columns), index=test_cat.index)
```

```python
train_cat_encoded_df.head()
```

```python
test_cat_encoded_df.head()
```

```python
X_train1 = pd.concat([train_num_scaled_df, train_cat_encoded_df], axis=1)
X_test1 = pd.concat([test_num_scaled_df, test_cat_encoded_df], axis=1)
```

```python
from sklearn.linear_model import LogisticRegression

classification = LogisticRegression(random_state =0, solver='lbfgs').fit(X_train1,y_train)
print(classification.score(X_train1, y_train))
print(classification.score(X_test1, y_test))
```

```python
predictions=classification.predict(X_test1)
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test,predictions)
#the 0's are correctly modelled but there is no result for Target B = 1
```

```python
from sklearn.utils import resample

y_train= y_train.reset_index(drop=True)
X_train1= X_train1.reset_index(drop=True)



train_data = pd.concat([X_train1, y_train], axis=1)
#X_train1.isna().sum()
y_train.tail(10)
```

```python
train_data.isna().sum()
```

```python
category_0 = train_data[train_data['TARGET_B'] == 0]
category_1 = train_data[train_data['TARGET_B'] == 1]



category_0_downsampled = resample(category_0,
                                  replace=False,
                                  n_samples=len(category_1),
                                  random_state=42)


resampled_down = pd.concat([category_0_downsampled, category_1], axis=0)

#resampled_down
```

```python
resampled_down.head()
```

```python
y_downsampled=resampled_down['TARGET_B']
X_downsampled=resampled_down.drop(columns=['TARGET_B'])
```

```python
classification_downsampled = LogisticRegression(random_state =0, solver='lbfgs').fit(X_downsampled,y_downsampled)
print(classification_downsampled.score(X_downsampled, y_downsampled))
```

```python
predictions_downsampled=classification_downsampled.predict(X_test1)
confusion_matrix(y_test,predictions_downsampled)
#predicted 1's but still has 7387 0 which should have been 1 and 428 1's which should have been 0
```

```python
#upsampling
category_1_oversampled = resample(category_1,
                                  replace=True,
                                  n_samples = len(category_0))
```

```python
print(category_0.shape)
print(category_1_oversampled.shape)
```

```python
data_upsampled = pd.concat([category_0, category_1_oversampled], axis=0)
```

```python
y_upsampled=data_upsampled['TARGET_B']
X_upsampled=data_upsampled.drop(columns=['TARGET_B'])
```

```python
classification_upsampled = LogisticRegression(random_state =0, solver='lbfgs').fit(X_upsampled,y_upsampled)
print(classification_upsampled.score(X_upsampled, y_upsampled))
```

```python
predictions_upsampled=classification_upsampled.predict(X_test1)
confusion_matrix(y_test,predictions_upsampled)
```

```python
data_upsampled = pd.concat([X_upsampled, y_upsampled], axis=1)
data_downsampled= pd.concat([X_downsampled, y_downsampled], axis=1)

data_upsampled.to_csv('data_upsampled', index=False)
data_downsampled.to_csv('data_downsampled', index=False)
```

```python
X_train_upsampled=data_upsampled.drop('TARGET_B',axis=1)
y_train_upsampled=data_upsampled['TARGET_B']
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=5, # max number of questions to ask
                             min_samples_split=20, # amount of rows still considered at every question
                             min_samples_leaf =20, # ultimate answer based on at least this many rows
                             max_samples=0.8, # fraction of X-train to use in each tree
                             random_state=42)
clf.fit(X_train_upsampled, y_train_upsampled)
print(clf.score(X_train_upsampled, y_train_upsampled))
print(clf.score(X_test1, y_test))

y_pred = clf.predict(X_test1)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix

clf = RandomForestClassifier(max_depth=10, # max number of questions to ask
                             min_samples_split=20, # amount of rows still considered at every question
                             min_samples_leaf =20, # ultimate answer based on at least this many rows
                             max_samples=0.8, # fraction of X-train to use in each tree
                             random_state=42)
clf.fit(X_train_upsampled, y_train_upsampled)
print(clf.score(X_train_upsampled, y_train_upsampled))
print(clf.score(X_test1, y_test))

y_pred = clf.predict(X_test1)
display(y_test.value_counts())
display(confusion_matrix(y_test, y_pred))
```

```python
all_data = pd.concat((numerical,categorical,targets),axis=1)
all_data.head()

donations_data = all_data[all_data['TARGET_B']==1]
X = donations_data.drop(columns=['TARGET_B','TARGET_D'])
y = donations_data['TARGET_D']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)

numericalX_train   = X_train.select_dtypes(np.number)
numericalX_test    = X_test.select_dtypes(np.number)
categoricalX_train = X_train.select_dtypes(object)
categoricalX_test  = X_test.select_dtypes(object)

# we OneHotEncode the categoricals so we can use the same dataset to perform a regression later (in the lab).
# it is not needed for a DecisionTree or RandomForest model
from sklearn.preprocessing import OneHotEncoder
encoder = OneHotEncoder(drop='first').fit(categoricalX_train)
encoded_categorical_train = encoder.transform(categoricalX_train).toarray()
encoded_categorical_train = pd.DataFrame(encoded_categorical_train, columns=encoder.get_feature_names_out()) # needed to avoid error
encoded_categorical_test = encoder.transform(categoricalX_test).toarray()
encoded_categorical_test = pd.DataFrame(encoded_categorical_test, columns=encoder.get_feature_names_out())

# we scale the numericals so we can use the same dataset to perform a regression later (in the lab).
# it is not needed for a DecisionTree or RandomForest model
from sklearn.preprocessing import MinMaxScaler
transformer = MinMaxScaler().fit(numericalX_train)
scaled_numerical_train = transformer.transform(numericalX_train)
scaled_numerical_train = pd.DataFrame(scaled_numerical_train, columns=numericalX_train.columns)
scaled_numerical_test  = transformer.transform(numericalX_test)
scaled_numerical_test  = pd.DataFrame(scaled_numerical_test, columns=numericalX_train.columns)

X_train_treated = pd.concat([scaled_numerical_train, encoded_categorical_train], axis = 1)
X_test_treated  = pd.concat([scaled_numerical_test,  encoded_categorical_test],  axis = 1)



X_train_treated.isna().sum()

X_test_treated.isna().sum()

from sklearn.linear_model import LinearRegression as LinReg

linreg=LinReg()    # model
linreg.fit(X_train_treated, y_train)   # model training
y_pred_linreg=linreg.predict(X_test_treated)   # model prediction

print ('train R2: {} -- test R2: {}'.format(linreg.score(X_train_treated, y_train),
                                            linreg.score(X_test_treated, y_test)))

from sklearn.metrics import mean_squared_error as mse


train_mse=mse(linreg.predict(X_train_treated), y_train)
test_mse=mse(linreg.predict(X_test_treated), y_test)

print ('train MSE: {} -- test MSE: {}'.format(train_mse, test_mse))

y_pred_linreg
#donations for the different X_test_treated

#from sklearn.decomposition import PCA
#pca = PCA(0.75)
#pca.fit(X_train_treated)

#pca.explained_variance_ratio_

from sklearn.tree import DecisionTreeRegressor
model1 = DecisionTreeRegressor()
from sklearn.linear_model import LinearRegression
model2 = LinearRegression()
from sklearn.neighbors import KNeighborsRegressor
model3 = KNeighborsRegressor()
from sklearn.model_selection import cross_val_score


model_pipeline = [model1, model2, model3]
model_names = ['Decision Tree Regressor', 'Linear Regression', 'KNN']
scores = {}
for model, model_name in zip(model_pipeline, model_names): #zip goes through elements of 2 lists at the same time
    mean_score = np.mean(cross_val_score(model, X_train_treated, y_train, cv=5))
    scores[model_name] = mean_score
print(scores)


#the linear regression is the best


numericalX    = X.select_dtypes(np.number)
categoricalX = X.select_dtypes(object)

encoded_categorical_X = encoder.transform(categoricalX).toarray()
encoded_categorical_X_df=pd.DataFrame(encoded_categorical_X, columns=encoder.get_feature_names_out())

scaled_numerical_X = transformer.transform(numericalX)
scaled_numerical_X_df= pd.DataFrame(scaled_numerical_X, columns=numericalX.columns)

X_df = pd.concat([scaled_numerical_X_df, encoded_categorical_X_df], axis = 1)

#X_df

```
